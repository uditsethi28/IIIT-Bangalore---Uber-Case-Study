```python
# loading libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the data from csv
df = pd.read_csv(r'C:\Users\admin4\Desktop\IIIT Bangalore - UPGRAD\Uber Request Data.csv')
```

```python
# DATA CLEANING
# Changing request timestamp to date time format
df['Request timestamp']=pd.to_datetime(df['Request timestamp'],dayfirst=True)
```

```python
# Changing drop timestamp to date time format
df['Drop timestamp']=pd.to_datetime(df['Drop timestamp'],dayfirst=True)
```

```python
df
```

```python
# 1.1 
#plot to visualise the frequency of requests
plot1=sns.countplot(x='Status',data=df)
```

```python
# 1.2
# plot to identify the most problematic types of requests
plot2=sns.countplot(x='Pickup point',hue='Status',data=df)
```

```python
# deriving hour from Request timestamp
df['Request Hour'] = df['Request timestamp'].dt.hour
```

```python
# deriving hour from drop hour
df['Drop Hour'] = df['Drop timestamp'].dt.hour
```

```python
df
```

```python
# deriving different time slots
df['Time Slot']=pd.cut(df['Request Hour'],bins=[0,6,12,18,24],labels=['Early Morning','Morning','Afternoon','Late Evening'],include_lowest=True,right=False)
```

```python
#  pickup point as city
city = df[df['Pickup point']=="City"]
```

```python
# extracting data of pickup point as Airport
airport = df[df['Pickup point']=="Airport"]
```

```python
# 1.3.1
# plot to identify the time slot and status at pickup point city
plot3=sns.countplot(x='Time Slot',hue='Status',data=city)
```

```python
# 1.3.2
# plot to identify the time slot and status at pickup point Airport
plot4=sns.countplot(x='Time Slot',hue='Status',data=airport)
```

```python
# 2
# create new data frame and group it by time slot and find demand
df2 = pd.DataFrame(df.groupby(by='Time Slot', as_index=False)['Request id'].count())
```

```python
# add and name the column as demand
df2.columns.values[1] = 'Demand'
```

```python
# create new data frame and group it by time slot and find supply
supply =df[df.Status == 'Trip Completed'].groupby(by='Time Slot', as_index=False)['Request id'].count()
```

```python
# name the column as supply
supply.columns.values[1] = 'Supply'
```

```python
# add column supply to df2 from supply
df2['Supply'] = supply['Supply']
```

```python
# add colum Gap as Demand - Supply
df2['Gap'] = df2['Demand']- df2['Supply']
```

```python
df2
```

```python
# 2.1
# plot for Gap against different time slots
plot5=sns.catplot(x='Time Slot',y='Gap',data=df2,kind='bar')
```

```python
# create new data frame and group it by time slot,pickup point and find demand
df3 = pd.DataFrame(df.groupby(by=['Time Slot','Pickup point'], as_index=False)['Request id'].count())
```

```python
# name and add the column as demand
df3.columns.values[2] = 'Demand'
```

```python
# name the column as supply
supply2.columns.values[2] = 'Supply'
```

```python
# add colum Gap as Demand - Supply
df3['Gap'] = df3['Demand']- df3['Supply']
```
